
PERFORM DATA CLEANING - REMOVE PUNCTUATION,NUMBERS,STOP WORDS FROM TEXT
AND PERFORM COUNT VECTORIZATION

```python
import string
import nltk
import re
from nltk.corpus import stopwords
string.punctuation
```

```python
nltk.download('wordnet')
```

```python
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
def remove_non_english_characters(message):
    encoded_string = message.encode("ascii", "ignore")
    return encoded_string.decode()
```

```python
#lemmatization
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
def done_lemmatization(message):
    result=[lemmatizer.lemmatize(word) for word in message.split()]
    return " ".join(result)
```

```python
# Convert more than 2 letter repetitions to 2 letter
# funnnnny --> funny
def remove_repetedLetters(message):
    message=[ re.sub(r'(.)\1+', r'\1\1', word) for word in message.split()]
    return ' '.join(message)
```

```python
def clean_tweet(message):
        tweet = message.lower() # convert text to lower-case
        tweet = re.sub("\d+", "", tweet) #remove number
        tweet = re.sub('((www\.[^\s]+)|(https?://[^\s]+))', 'URL', tweet) # remove URLs
        tweet = re.sub('@[^\s]+', 'AT_USER', tweet) # replace usernames with 'AT_USER'
        tweet = re.sub(r'#([^\s]+)', r'\1', tweet) # remove the # in #hashtag
        tweet = remove_repetedLetters(tweet)
        return tweet
```

```python
def message_cleaning(message):
    message=remove_non_english_characters(message)
    message=done_lemmatization(message)
    message=clean_tweet(message)
    punc_removed = [char for char in message if char not in string.punctuation]
    punc_removed_join=''.join(punc_removed)
    clean_text=[word for word in punc_removed_join.split() if len(word)>2]
    return clean_text
```

```python
tweets_df_clean = tweets_df['tweet'].apply(message_cleaning)
```

```python
print(tweets_df_clean[1]) # show the cleaned up version
```

```python
print(tweets_df['tweet'][1]) # show the original version
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(analyzer = message_cleaning, dtype = 'uint8')
tweets_countvectorizer = vectorizer.fit(tweets_df['tweet'])
tweets_countvectorizer_transformed=tweets_countvectorizer.transform(tweets_df['tweet'])
```

```python
print(tweets_countvectorizer.get_feature_names())
```

```python
print(tweets_countvectorizer_transformed.toarray())  
```

```python
tweets_countvectorizer_transformed.shape
```

```python
header_list_X = tweets_countvectorizer.get_feature_names()
tweets = pd.DataFrame(tweets_countvectorizer_transformed.toarray(),columns = header_list_X)
```

```python
X = tweets
X
```

```python
y = tweets_df['label']
y
```


TRAIN-TEST SPLIT


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35)
```


TRAIN A NAIVE BAYES CLASSIFIER MODEL


```python
from sklearn.naive_bayes import MultinomialNB

NB_classifier = MultinomialNB()
NB_classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
y_predict_test = NB_classifier.predict(X_test)
cm = confusion_matrix(y_test, y_predict_test)

print(cm)
```

```python
print(classification_report(y_test, y_predict_test))
```

SCORE PREDICT FOR NAIVE BAYES

```python
NB_classifier.score(X_test, y_test)
```

```python
accuracy_score(y_test, y_predict_test)
```


LOGISTIC REGRESSION


```python
from sklearn.linear_model import LogisticRegression

# Train the model
Logistic_model = LogisticRegression()
Logistic_model.fit(X_train, y_train)
```

```python
y_predict_test_logisticRegression = Logistic_model.predict(X_test)
cm = confusion_matrix(y_test, y_predict_test_logisticRegression)

print(cm)
```

```python
print(classification_report(y_test, y_predict_test_logisticRegression))
```

```python
accuracy_score(y_test, y_predict_test_logisticRegression)
```


SVM MACHINE


```python
from sklearn.svm import LinearSVC
# Train the model
LinearSVC_model = LinearSVC()
LinearSVC_model.fit(X_train, y_train)
```

```python
y_predict_test_LinearSVC = LinearSVC_model.predict(X_test)
cm = confusion_matrix(y_test, y_predict_test_LinearSVC)

print(cm)
```

```python
print(classification_report(y_test, y_predict_test_LinearSVC))
```

```python
accuracy_score(y_test, y_predict_test_LinearSVC)
```
